```python
#BY OBSERVING HEATMAP IT SUGGEST TO DROP SOME COLUMNS WHICH VERY LESS RELAVENT TO US 
ftrain=train.drop(['Col3','Col4','Col5','Col6','Col7','Col8','Col9','Col11','Col2391','Col2395','Col2396','Col2397'], axis=1)
ftrain
```

```python
#FINAL TRAIN DATASET FOR MODEL SELECTION
x=ftrain.drop(['Col2','Col1'],axis=1)
Y_train=ftrain['Col2']
```

```python
#FINAL TEST DATASET FOR MODEL SELECTION
ftest=test.drop(['Col1','Col3','Col4','Col5','Col6','Col7','Col8','Col9','Col11','Col2391','Col2395','Col2396','Col2397'], axis=1)
```

```python
#FEATURE SCALING OF TRAIN AND TEST DATASET FOR BETTER PREDICTION
```

```python

from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(x)
X_test = sc.fit_transform(ftest)
X_test
```

```python
# MODEL SELECTION--> WE ARE CHECKING CONDITION FOR EITHER YES OR NO,SO IT'S BASICALLY 0,1
# MEANS WE HAVE TO SELECT CLASSIFICATION MODEL,IN THIS CASE WE WILL SELECT KNN CLASSIFIER 
```

```python
from sklearn.neighbors import KNeighborsClassifier
model=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
model.fit(X_train,Y_train)
```

```python
#PREDICTING TEST DATA
y_pred=model.predict(X_test)
```

```python
y_pred
```

```python
# MODEL PERFORMANCE PERCENTAGE
model.score(X_train,Y_train)
```

```python
# IN PREVIOUS TEST DATA WE HAVE NO INTIAL VALUE OF TARGET VARIABLES ,WE HAVE JUST PREDICTED IT,TO KNOW CONFUSION MATRIX WE NEED 
# WE NEED GIVEN TARGET VARIABLE AND PREDICTED TARGET VARIABLE ,SO FOR DOING THIS WE JUST SPLIT TRAIN DATA IN TRAIN AND TEST DATA
#SET TO PERFORM CONFUSION MATRIX.
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,Y_train,test_size=0.2)
```

```python


sc = StandardScaler()
X_train1 = sc.fit_transform(x_train)
X_test2 = sc.fit_transform(x_test)

X_test2
```

```python

model.fit(X_train1,y_train)
```

```python
y_pred2=model.predict(X_test2)
```

```python
y_pred2
```

```python
model.score(X_train1,y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred2)
cm
```
